```python
from math import radians, cos, sin, asin, sqrt, isnan
#Función para las distancias geográficas. (Haversine formula)


def distance(p1_lat, p1_lon, p2_lat, p2_lon):
     
  
    
    lon1 = radians(p1_lon)
    lon2 = radians(p2_lon)
    lat1 = radians(p1_lat)
    lat2 = radians(p2_lat)
      
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
 
    c = 2 * asin(sqrt(a))   
   
    r = 6371 #radio de la tierra aproximado
      

    return(c * r)
```

```python

max_distance=0.00017986

df["group_diff"] = df.sort_values("latitude")["latitude"]\
                     .diff()\
                     .gt(max_distance)\
                     .cumsum()

df[["latitude2","longitude2"]]=df.sort_values(["group_diff","longitude"])[["latitude","longitude"]].shift()

l=lambda row: distance(row["latitude"],row["longitude"],row["latitude2"],row["longitude2"])

df["dist"]=df.sort_values(by=['group_diff','latitude']).apply(l,axis=1)


    
df["group_diff2"] = df.sort_values(["group_diff","longitude"])["dist"]\
                     .gt(0.017)\
                     .cumsum()


df.sort_values(["group_diff","longitude"])

with pd.option_context('display.max_rows', None,
                       'display.max_columns', 10,
                       'display.precision', 10,
                       ):
    print(df.sort_values(["group_diff","longitude"])[["vehicle","latitude","longitude","timestamp","group_diff","dist","group_diff2"]])
```
